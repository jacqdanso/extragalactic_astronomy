```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.cosmology import FlatLambdaCDM
import seaborn as sea
import scipy.constants as const
%matplotlib inline
```

```python
# set plotting parameters
plt.rc('text', usetex=True)
sea.set(style='white')
sea.set(style='ticks')
sea.set_style({'xtick.direct'
               'ion': 'in','xtick.top':True,'xtick.minor.visible': True,
               'ytick.direction': "in",'ytick.right': True,'ytick.minor.visible': True})
```

```python
plt.figure(figsize=(15,10))
plt.gca().invert_xaxis()
redshifts = np.linspace(0, 15, num=1000)
model = FlatLambdaCDM(H0=67.36, Om0=0.3153, Tcmb0=2.725, Neff=2.99)
plt.plot(redshifts, model.Om(redshifts), label=r'$\Omega_m$', lw = 2);
plt.plot(redshifts, model.Ode(redshifts), label=r'$\Omega_\Lambda$', lw = 2, c='magenta');
plt.plot(redshifts, model.Ogamma(redshifts)+model.Onu(redshifts), label=r'$\Omega_{rad}$', lw = 2 ,ls = 'dashed',\
        c='brown')
plt.legend(loc='best');
plt.xlabel(r'$z$')
plt.ylabel(r'$\Omega_{parameter}$')
plt.savefig('omega_evolution.png')
```
